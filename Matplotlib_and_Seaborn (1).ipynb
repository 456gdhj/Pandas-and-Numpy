```python
import pandas as pd
df_sales=pd.read_excel("linechart.xlsx")
df_sales.head()
```

```python
from matplotlib import pyplot as plt
plt.figure(figsize=(12,4))
plt.plot(df_sales["Quarter"],df_sales["Fridge"])
```

```python
plt.plot(df_sales["Quarter"],df_sales["Fridge"],color="blue")
plt.plot(df_sales["Quarter"],df_sales["Dishwasher"],color="green")
plt.plot(df_sales["Quarter"],df_sales["Washing Machine"],color="orange")

plt.title("Product Sales")
plt.ylabel("Revenue(mln $)")
plt.xlabel("Financial Quarters")

plt.legend(["Fridge","Dishwasher","Washing Machine"])
```

```python
total_sales=df_sales[["Fridge","Dishwasher","Washing Machine"]].sum()
total_sales
```

```python
total_sales.index
```

```python
plt.pie(total_sales,labels=total_sales.index)
plt.show()
```

```python
plt.pie(total_sales,labels=total_sales.index,autopct="%1.1f%%")
plt.show()
```

```python
plt.pie(total_sales,labels=total_sales.index,autopct="%1.1f%%",explode=(0,1,0,),shadow=True,startangle=180)
plt.show()
```

```python
df_sales.plot(kind="bar",x="Quarter")
plt.xticks(rotation=45)
```
